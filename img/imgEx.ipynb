```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt 

#img = np.asarray(Image.open('./testimg.png'))
color_img  = Image.open('./fruits.jpg')
#color_img  = Image.open('./fruits.jpg').convert('L')

print(np.shape(color_img))
plt.imshow(color_img)
plt.show()
```

```python
imgArr = np.asarray(color_img)
print(np.shape(imgArr))
print(imgArr)
```

```python
color_pixel = np.array(color_img)
print(color_pixel.shape)
```

```python
gray_pixel = color_pixel.copy() #깊은 복사 
print(id(gray_pixel))
print(id(color_pixel))


for y  in range(gray_pixel.shape[0]):
    for x in range(gray_pixel.shape[1]):
        gray_pixel[y,x] = np.mean(gray_pixel[y,x])
#G에 높은 가중치를 주고, 흑백 사진을 보면 더 선명해짐 
        
plt.imshow(gray_pixel)
plt.show()

plt.imshow(color_pixel)
plt.show()
        
```

```python
#흑백 이미지를 2차원으로 표현 
#차원을 줄이고, 이미지 사이즈를 줄이고, convolution하면서 특징들만 뽑고 
gray_2d_pixel = gray_pixel[:,:,0] #R 값만 가져오고
print(np.shape(gray_2d_pixel))
print(gray_2d_pixel.shape)
print(gray_2d_pixel)
```
